```python
import gym
l=[]
k=[]
num_of_epi=1000
num_timesteps = 50
for i in range(num_of_epi):
  env = gym.make("FrozenLake-v0")
  state = env.reset()
  print('Time Step 0 :')
  print(' {}th episode'.format(i))
  #env.render()

  for t in range(num_timesteps):
    random_action = env.action_space.sample()
    new_state, reward, done, info = env.step(random_action)
    #print ('Time Step {} :'.format(t+1))
    l.append(reward)
    
    #env.render()
    if done:
      print(max(set(l)))
      k.append(max(set(l)))
      break
      
```

```python
plt.figure(figsize=(20,20))
x=np.linspace(1,1000,1000)
plt.scatter(x=x,y=k)
plt.xlabel("Episodes")
plt.ylabel("Reward")
plt.title("Episodes v/s reward for frozen lake")
```
